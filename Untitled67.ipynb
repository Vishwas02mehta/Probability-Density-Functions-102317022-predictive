```python



import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import minimize




df = pd.read_csv("/content/data.csv", encoding="latin1")

print("Columns in dataset:")
print(df.columns)




no2_cols = [col for col in df.columns if "no2" in col.lower()]

if len(no2_cols) == 0:
    raise ValueError("NO2 column not found in dataset!")

no2_col = no2_cols[0]
print("Using NO2 column:", no2_col)

x = df[no2_col].dropna().values.astype(float)




r = 102317022   # ðŸ”´ REPLACE WITH YOUR UNIVERSITY ROLL NUMBER

ar = 0.05 * (r % 7)
br = 0.3 * (r % 5 + 1)

def transform(x):
    return x + ar * np.sin(br * x)

z = transform(x)

def pdf(z, lam, mu, c):
    return c * np.exp(-lam * (z - mu)**2)

# Negative Log Likelihood
def negative_log_likelihood(params):
    lam, mu, c = params
    p = pdf(z, lam, mu, c)
    p = np.clip(p, 1e-9, None)
    return -np.sum(np.log(p))


initial_params = [0.01, np.mean(z), 1.0]
bounds = [(1e-6, None), (None, None), (1e-6, None)]

result = minimize(
    negative_log_likelihood,
    initial_params,
    bounds=bounds
)

lam_hat, mu_hat, c_hat = result.x


print("\n===== Estimated PDF Parameters =====")
print(f"Î» (lambda) = {lam_hat}")
print(f"Î¼ (mu)     = {mu_hat}")
print(f"c          = {c_hat}")


z_sorted = np.sort(z)
pz = pdf(z_sorted, lam_hat, mu_hat, c_hat)

plt.figure(figsize=(8, 5))
plt.hist(z, bins=50, density=True, alpha=0.5, label="Empirical z")
plt.plot(z_sorted, pz, color="red", linewidth=2, label="Learned PDF")
plt.xlabel("z")
plt.ylabel("Probability Density")
plt.legend()
plt.title("Learned Probability Density Function")
plt.show()
```
